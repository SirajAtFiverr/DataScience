## Finding Relationship

```python
corr_data = data.corr()
corr_data
```

```python
sns.heatmap(corr_data, annot=True)
plt.show()
```

```python
sns.relplot(x='age', y='fare', hue='sex', data=data)
plt.show()
```

```python
sns.catplot(x='sex', y='age', data=data, kind='box')
```

```python
data['fare_log'] = np.log(data['fare'])
```

```python
sns.boxplot(data['fare'])
plt.show()
```
