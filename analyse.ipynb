# Importing Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### READING DATASET

```python
data = pd.read_csv('rainfall.csv')
data.head()
```

```python
data.describe()
```

```python
data_2016 = data[data['Year'] == 2016]
data_2016
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
```

```python
sns.barplot(x='Month', y='Rainfall - (MM)', data=data_2016)
plt.title('RainFall in 2016')
plt.show()
```

```python
data_1901 = data[data['Year'] == 1901]
data_1901
```

```python
sns.barplot(x='Month', y='Rainfall - (MM)', data=data_1901)
plt.title('RainFall in 1902')
plt.show()
```

```python
data_2016_and_1901 = pd.concat((data_1901, data_2016))
data_2016_and_1901
```

```python
sns.barplot(x='Month', y='Rainfall - (MM)', hue='Year', data=data_2016_and_1901)
plt.title('RainFall Comparison')
plt.show()
```

```python
data = pd.read_csv('food.csv')
data.head()
```

```python
white_flour_data = data[data['cmname'] == 'Wheat flour - Retail'] 
white_flour_data.tail()
```

```python
date = map(lambda x: x.split('/')[-1], white_flour_data['date'].to_numpy())
white_flour_data['year'] = np.array(list(date))
```

```python
sns.lineplot(data=white_flour_data, x="year", y="price", hue='admname')
plt.title('White Flour Price in Pakistan')
plt.show()
```
