**Population:** (Whole Data)
- Population represets every single unit of a subject.
- A descriptive measure for an entire a Population is is its Parameter (Population mean)
- Parameters determine how distribution fits population
- The above dataset as a whole is popluation

**Sample** (A Subset of Population)
- A sepcific group that you will collect from Population
- Sample size is always less than Population Size
-  A descriptive measure for a sample is called Satistics (Sample mean)
-  We can use **Sample Satistics** to predict **Population Paramters**
-  Samples with more data produce more accurate results
-  We can take random sample by: `df.sample(250)`

We can see that data of **fare** is **postively-skewed**

### Calculating IQR of **fare**

```python
Q1 = df['fare'].quantile(0.25)
print("Q1:", Q1)

Q3 = df['fare'].quantile(.75)
print("Q3:", Q3)

IQR = Q3 - Q1
print("IQR:", IQR)
```

### Selecting Outliers of **fare**

```python
outliers = df['fare'][ (df['fare'] < (Q1 - 1.5 * IQR)) |
                      (df['fare'] > (Q3 + 1.5 * IQR))]
outliers
```
